# **visualization**

```python

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# **scatter plot**


```python
sns.set_context("paper")
sns.scatterplot(x='Category', y='Age', data=df, hue='Sex')
```

```python
sns.set_context("poster")
sns.relplot(data=df, x="Age", y="Sex", hue="Category")
```

```python
sns.set_context("poster")
sns.relplot(data=df, x="Age", y="Sex", hue="Category",col="ALB",col_wrap=5)
```

# **pair plot**

```python
sns.set_style("ticks")
sns.pairplot(df,hue = 'Sex',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```

```python
sns.set_context("poster")
sns.catplot(x="Category", col = 'Sex', data=df, kind = 'count', palette='husl')
plt.gcf().set_size_inches(70, 10)
plt.show()
```

```python
sns.set_context("poster")
sns.catplot(x="Age", col = 'Category', data=df, kind = 'count', palette='husl')
plt.gcf().set_size_inches(40, 10)
plt.show()
```
